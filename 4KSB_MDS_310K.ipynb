## 3. Caja de agua

```python
!/home/pecosa/gromacs-2022.3/build-gpu/bin/gmx editconf -f conf.gro -o newbox.gro -c -d 1.0 -bt cubic
```

```python
!/home/pecosa/gromacs-2022.3/build-gpu/bin/gmx solvate -cp newbox.gro -cs tip4p.gro -o solv.gro -p topol.top
```

## 6. EQUILIBRIO 1 (NVT)

Necesitamos 3 cosas:\
(1) Primero requerimos crear un archivo de índice: index.ndx \
gmx make_ndx -f em.gro -o index.ndx
(puedes generarlo localmente)
Para 150 mM, se selecionaron: \
"1" (protein)\
"13" (SOL)\
"16" (NA)\
"17" (CL)\
"q" para salir 

(2) Se descargó el archivo nvt.mdp \
(3) En la fila tc-grps se editó, colocando los archivos que llamaremos:\
tc-grps                 = Protein Non-Protein   ; two coupling groups - more accurate \
ref_t                   = 310     310           ; reference temperature, one for each group, in K\
gen_temp                = 310       ; temperature for Maxwell distribution

El index lo cree en la misma terminal porque el siguiente comando no se podía ejecutar por aquí:
(printf "1" | printf "13" | printf "16" | printf "17" | printf "q" ) | gmx make_ndx -f em.gro -o index.ndx

ojo que gmx se colocó con el path: /home/pecosa/gromacs-2022.3/build-gpu/bin/gmx  

Luego de disponer del index, se crea el tpr para NVT

```python
!/home/pecosa/gromacs-2022.3/build-gpu/bin/gmx grompp -f nvt.mdp -c em.gro -r em.gro -p topol.top -n index.ndx -o nvt.tpr
```
